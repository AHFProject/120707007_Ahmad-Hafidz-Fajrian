```python
#Import library
import numpy as np  #import lib numpy
import imageio     #import lib imageio
import matplotlib.pyplot as plt #import lib matplotlib
```

```python
img = imageio.imread("asimo.jpg") #load gambar asimo
```

```python
#Resolusi dan Tipe Gambar 
img_height = img.shape[0]    #Tinggi gambar 
img_width = img.shape[1]     # Lebar gambar
img_channel = img.shape[2]   #Channel Gambar
img_type = img.dtype         #tioe gambar 
```

```python
img_brightness = np.zeros(img.shape, dtype=np.uint8) #kecerahan gryscale
```

```python
# Melakukan penambahan brightness dengan nilai yg menjadi parameter
def brighter(nilai):   #mendefinisi nilai Brighter 
    for y in range(0, img_height):  # ukuran Tinggi 
        for x in range(0, img_width): #ukuran Lebar 
            red = img[y][x][0]     #warna Merah
            green = img[y][x][1]   #warna hijau
            blue = img[y][x][2]    #warna biru
            gray = (int(red) + int(green) + int(blue)) / 3 #warna RGB di bagi 3
            gray += nilai  
            if gray > 255: #jika nilai gray lebih dari 255
                gray = 255 # nilaigray samadengan 255
            if gray < 0: #jika nilai gray kurang dari 0
                gray = 0 #maka nilai gray sama dengan 0
            img_brightness[y][x] = (gray, gray, gray)
```

```python
brighter(-100) #jika nilai brighter -100
plt.imshow(img_brightness) #menampilkan img brighness
plt.title("Brightness -100") #nampilin text  
plt.show() #tampilkan 

brighter(100)  #jika nilai brighter 100
plt.imshow(img_brightness)  #menampilkan img brighness
plt.title("Brightness 100") #nampilin text 
plt.show()  #tampilkan 
```

```python
img_rgbbrightness = np.zeros(img.shape, dtype=np.uint8)
```

```python
def rgbbrighter(nilai): #definisi nilai rgbbrighter
    for y in range(0, img_height): #tinggi gambar
        for x in range(0, img_width): #lebar gambar 
            red = img[y][x][0] #gambar merah
            red += nilai 
            if red > 255: #jika nilai merah  lebih dari 255
                red = 255 #warna merah nilainya sama dengan 255
            if red < 0: #jika warna merah nilainya kurang dari 0
                red = 0 #maka merah nilainya 0
            green = img[y][x][1] #gambar warna merah
            green += nilai
            if green > 255:    #jika nilai hijau  lebih dari 255
                green = 255    #warna merah nilainya sama dengan 255
            if green < 0:      #jika hijau nilainya kurang dari 0
                green = 0      #maka hijau nilainya sama dengan 0
            blue = img[y][x][2] #gambar warna biru
            blue += nilai
            if blue > 255:     #jika  nilai biru lebih dari 255
                blue = 255    # maka biru nilainya 255
            if blue < 0:      #jika biru nilainya kurang dari 0
                blue = 0      # maka nilai biru sama dengan 0
            img_rgbbrightness[y][x] = (red, green, blue)
```

```python
rgbbrighter(-100)              #rgbbrighter nilainya -100
plt.imshow(img_rgbbrightness)  #menampilkan img rgb brightness
plt.title("Brightness -100")   # text aja 
plt.show()                     #menampilkan perintah

rgbbrighter(100)               #rgbbrighter nilainya 100
plt.imshow(img_rgbbrightness)  #menampilkan img rgb brightness
plt.title("Brightness 100")    # text aja 
plt.show()                     #menampilkan perintah
```

```python
img_contrass = np.zeros(img.shape, dtype=np.uint8) #contrass gambar
```

```python
def contrass(nilai):                    #definisi nilai kontrass
    for y in range(0, img_height):      #tinggi gambar
        for x in range(0, img_width):   #lebar gambar
            red = img[y][x][0]          #merah
            green = img[y][x][1]        #hijau
            blue = img[y][x][2]         #biru
            gray = (int(red) + int(green) + int(blue)) / 3  #Nilai RGB bagi 3
            gray *= nilai
            if gray > 255:  # jika nilai gray lebih dari 255 
                gray = 255  # nilai gray sama dengan 255
            img_contrass[y][x] = (gray, gray, gray)  #contrasnya gray semua
```

```python
contrass(2) #jika nilai kontrasnya 2
plt.imshow(img_contrass)  #menampilkan img contrass
plt.title("Contrass 2")  #text aja
plt.show()    #menampilkan 

contrass(3)   #jika nilai kontrasnya 3
plt.imshow(img_contrass)  #menampilkan img contrass
plt.title("Contrass 3")  #text aja
plt.show()    # menampilkan 
```

```python
img_autocontrass = np.zeros(img.shape, dtype=np.uint8)
```

```python
def autocontrass():
    xmax = 300
    xmin = 0
    d = 0
    # Mendapatkan nilai d, dimana nilai d ini akan berpengaruh pada hitungan
    # untuk mendapatkan tingkat kontras
    for y in range(0, img_height):  #tinggi gambar
        for x in range(0, img_width): #lebar gambar
            red = img[y][x][0]     #merah
            green = img[y][x][1]   #hijau 
            blue = img[y][x][2]    #biru
            gray = (int(red) + int(green) + int(blue)) / 3 #Nilai RGB di bagi  3
            if gray < xmax: #jika gray kurang dari nilai xmax
                xmax = gray  # maka xmax = gray
            if gray > xmin:   # jika gray lebih dari xmin
                xmin = gray   #maka xmin sama dengan gray
    d = xmin-xmax
    for y in range(0, img_height):    #tinggi gambar
        for x in range(0, img_width): #lebar gambar
            red = img[y][x][0]      #merah
            green = img[y][x][1]    #hijau 
            blue = img[y][x][2]     #biru
            gray = (int(red) + int(green) + int(blue)) / 3  #Nilai RGB di bagi  3
            gray = int(float(255/d) * (gray-xmax))
            img_autocontrass[y][x] = (gray, gray, gray)
```

```python
autocontrass()   #auto kontrass 
plt.imshow(img_autocontrass)  #menampilkan img auto kontras
plt.title("Contrass Autolevel")  #text aja
plt.show()  # menampilkan 
```
