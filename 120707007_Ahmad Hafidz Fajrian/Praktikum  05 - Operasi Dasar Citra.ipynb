```python
#import library
import matplotlib.pyplot as plt  #import lib matplot lib
%matplotlib inline
from PIL import Image    
from skimage import data
from skimage.io import imread
from skimage.color import rgb2gray 
from skimage.util import invert
from skimage import util

import numpy as np
```

```python
# membuka gambar asimo.jpg
astronautImage = Image.open('asimo.jpg')

# memotong bagian tertentu dari gambar asimo
astroCropped = astronautImage.crop((64, 0, 320, 256))

# membuka gambar dari file Hero.jpg
cameraImage = Image.open('Hero.jpg')

# memotong bagian tertentu dari gambar Hero
cameraCropped = cameraImage.crop((128, 64, 320, 256))

# mencetak bentuk gambar sebelum dan sesudah dipotong
print('Astro Ori Shape : ',astronautImage.size)
print('Astro Crop Shape : ',astroCropped.size)
print('Camera Ori Shape : ',cameraImage.size)
print('Camera Crop Shape : ',cameraCropped.size)

# menampilkan kedua gambar asli dan yang sudah dipotong
fig, axes = plt.subplots(2, 2, figsize=(12, 12))
ax = axes.ravel()

ax[0].imshow(astronautImage)
ax[0].set_title("Citra Input 1")

ax[1].imshow(cameraImage)
ax[1].set_title('Citra Input 2')

ax[2].imshow(astroCropped)
ax[2].set_title("Citra Output 1")

ax[3].imshow(cameraCropped)
ax[3].set_title('Citra Output 2')
```

```python
astroCropped = np.asarray(astroCropped)  

inv = util.invert(astroCropped) #invert gambar asimo
print('Shape Input : ', astroCropped.shape) 
print('Shape Output : ',inv.shape)

fig, axes = plt.subplots(2, 2, figsize=(12, 12))
ax = axes.ravel()

ax[0].imshow(astroCropped)    #menampilkan gambar asimo
ax[0].set_title("Citra Input")

ax[1].hist(astroCropped.ravel(), bins=256) #histogram gambar asimo
ax[1].set_title('Histogram Input')

ax[2].imshow(inv)  #invert gambar asimo
ax[2].set_title('Citra Output (Inverted Image)')

ax[3].hist(inv.ravel(), bins=256)  #histogram gambar asimo invert
ax[3].set_title('Histogram Output')
```

```python
cameraCropped = np.asarray(cameraCropped)

inve = util.invert(cameraCropped)   #invert gambar bighero
print('Shape Input : ', cameraCropped.shape)
print('Shape Output : ',inve.shape)

fig, axes = plt.subplots(2, 2, figsize=(12, 12))
ax = axes.ravel()

ax[0].imshow(cameraCropped)    #menampilkan gambar big hero
ax[0].set_title("Citra Input")

ax[1].hist(cameraCropped.ravel(), bins=256) #menampilkan hitogram bighero
ax[1].set_title('Histogram Input')

ax[2].imshow(inve)             #menampilkan gambar bighero invert
ax[2].set_title('Citra Output (Inverted Image)')

ax[3].hist(inve.ravel(), bins=256) #menampilkan histogram bighero
ax[3].set_title('Histogram Output')
```
